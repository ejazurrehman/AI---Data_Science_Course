## Nested Loops in Python

- A nested loop means placing one loop inside another loop. The inner loop runs completely for each single iteration of the outer loop.
- In Python, any loop (for or while) can be nested inside another loop.

### Syntax

for outer_variable in outer_sequence:
    for inner_variable in inner_sequence:
        # Code block


#### Or using while loops:

while condition1:
    while condition2:
        # Code block


### Example 1 – Nested for loops

```python
# Printing a multiplication table (1 to 3)
for i in range(1, 4):  # Outer loop
    for j in range(1, 4):  # Inner loop
        print(f"{i} x {j} = {i*j}")
    print("-----")  # Separator after each row
```

### Example 2 – Nested while loop

```python
i = 1
while i <= 3:  # Outer loop
    j = 1
    while j <= 3:  # Inner loop
        print(f"i={i}, j={j}")
        j += 1
    i += 1
```

### Example 3 – Combining for and while

```python
for i in range(1, 4):  # Outer loop
    j = 1
    while j <= 2:  # Inner loop
        print(f"Outer={i}, Inner={j}")
        j += 1
```

### Key Points:

- Inner loop runs fully for each single outer loop iteration.
- Can be for inside for, while inside while, or for inside while and vice versa.
- Often used for working with matrices, patterns, and combinations.
