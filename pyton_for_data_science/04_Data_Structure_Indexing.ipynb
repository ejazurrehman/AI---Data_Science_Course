## Data Structure & Indexing

### 1. Data Structure in Python
- A data structure is a way of organizing, storing, and managing data so it can be accessed and modified efficiently.
- Python has built-in data structures, which are easy to use without extra setup, and user-defined data structures that you can create yourself.

### Common Built-in Data Structures in Python

| Data Structure | Description                          | Example                                |
| -------------- | ------------------------------------ | -------------------------------------- |
| **List**       | Ordered, mutable collection of items | `my_list = [10, 20, 30]`               |
| **Tuple**      | Ordered, immutable collection        | `my_tuple = (10, 20, 30)`              |
| **Set**        | Unordered, unique elements           | `my_set = {10, 20, 30}`                |
| **Dictionary** | Key-value pairs                      | `my_dict = {"name": "Ali", "age": 25}` |


```python
# List (mutable)
fruits = ["apple", "banana", "mango"]
fruits.append("orange")  # Add an item
print(fruits)  # ['apple', 'banana', 'mango', 'orange']

# Tuple (immutable)
coordinates = (10.5, 20.8)
print(coordinates[0])  # 10.5

# Set (unique values, unordered)
unique_numbers = {1, 2, 3, 3}
print(unique_numbers)  # {1, 2, 3}

# Dictionary (key-value pairs)
student = {"name": "Aisha", "age": 22}
print(student["name"])  # Aisha
```

### 2. Indexing in Python
- Indexing means accessing elements of a data structure by their position (index).
- Python uses zero-based indexing:
  - First element → index 0
  - Second element → index 1
  - Last element → index -1 (negative indexing works too)

```python
# Indexing in List
numbers = [10, 20, 30, 40, 50]
print(numbers[0])   # 10 (first element)
print(numbers[-1])  # 50 (last element)

# Indexing in String
word = "Python"
print(word[0])    # 'P'
print(word[-1])   # 'n'

# Indexing in Tuple
my_tuple = ("red", "green", "blue")
print(my_tuple[1])  # green

# Indexing in Dictionary (via keys)
person = {"name": "Sara", "age": 30}
print(person["age"])  # 30
```
